```python
preprocessed_text = re.split(r'([,.:;?_!"()\']|--|\s)', text)

preprocessed_text = [item for item in preprocessed_text if item.strip()]

print(preprocessed_text[:100])
print(len(preprocessed_text))
```

```python
ordered_text = sorted(set(preprocessed_text))
print(ordered_text[:100])

print(len(ordered_text))
```

```python
vocab = {
    token : id for id,token in enumerate(ordered_text)
}
```

```python
for id, item in enumerate(vocab.items()):
  print(id)
  if id > 10:
    break
```

```python
class SimpleTokenizerV1:
  def __init__(self,  vocab):
    self.str_to_int = vocab
    self.int_to_str = { i : s for s, i in vocab.items()}

  def encode(self, text):
    preprocessed_text = re.split(r'([,.:;?_!"()\']|--|\s)', text)
    preprocessed_text = [text for text in preprocessed_text if text.strip()]
    return [ self.str_to_int[s] for s in preprocessed_text]

  def decode(self, ids):
    #print(self.int_to_str)
    text = " ".join([self.int_to_str[id] for id in ids])
    #removes unnecessary spaces before certain punctuation marks, making the decoded text look natural (e.g., "Hello , world !" → "Hello, world!").
    #one or more spaces followed by a punctuation mark
    #the whole thing with just the punctuation mark
    text = re.sub(r'\s+([,.?!"()\'])', r'\1', text)
    return text


Tokenizer = SimpleTokenizerV1(vocab)
text = """"I It's the last he painted, you know,"
           Mrs. Gisburn said with pardonable pride."""

ids = Tokenizer.encode(text)
decoded_text = Tokenizer.decode(ids)
print(ids,decoded_text)
```

# Handling Unknown Token (unk token)
# Also add 'endoftext' token after between differen sources of technique.

```python
ordered_text.extend(["<|endoftext|>", "<|unk|>"])
```

```python
print(ordered_text[-5:])
```

```python
vocab = {
    token : id for id,token in enumerate(ordered_text)

}
```

```python
class SimpleTokenizerV2:
  def __init__(self,vocab):
    self.str_to_int = vocab
    self.int_to_str = { i: s for s, i in vocab.items()}
  def encode(self, text):
    preprocessed_text =  re.split(r'([,.:;?_!"()\']|--|\s)', text)
    preprocessed_text = [text for text in preprocessed_text if text.strip()]
    preprocessed_text = [
        item if item in self.str_to_int
        else "<|unk|>"
        for item in preprocessed_text
    ]
    ids = [self.str_to_int[s] for s in preprocessed_text]
    return ids
  def decode(self, ids):
    text = " ".join([self.int_to_str[id] for id in ids])
    text = re.sub(r'\s+([,.:;?!"()\'])', r'\1', text)
    return text


tokenizer = SimpleTokenizerV2(vocab)
```

```python
text1 = "Hello world!! How are you, Python is Awesome!"
text2 = "Thank you very much"
join_text = "<|endoftext|>".join([text1,text2])
print(join_text)
ids = tokenizer.encode(join_text)
d_text = tokenizer.decode(ids)
print(d_text)
```
