```python
import tensorflow_hub as hub
import tensorflow_text as text
import pandas as pd
from sklearn.utils import resample
import tensorflow as tf
```

```python
df = pd.read_csv("/content/spam.csv")
df.head()
```

```python
df['Category'] = df['Category'].map({'spam':1,'ham':0})
df.head()
```

```python
df['Category'].value_counts()
```

```python
df_majority =df[df.Category==0]
df_minority =df[df.Category==1]

df_minority_upsampled = resample(
                                 df_minority,
                                 replace= True,
                                 n_samples = len(df_majority),
                                 random_state = 42
)

df_balanced = pd.concat([df_majority,df_minority_upsampled])
df_balanced.head()
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(df_balanced['Message'],df_balanced['Category'], stratify=df_balanced['Category'])


```

```python
X_train[:10]
```

```python
from sklearn.feature_extraction.text import CountVectorizer
v = CountVectorizer()
X_train_count = v.fit_transform(X_train.values)
X_train_count
```

```python
X_train_np = X_train_count.toarray()
X_train_np.shape
```

```python
from sklearn.naive_bayes import MultinomialNB

model = MultinomialNB()
model.fit(X_train_count, y_train)
```

```python
X_test_cv = v.transform(X_test)
```

```python
from sklearn.metrics import classification_report

y_pred = model.predict(X_test_cv)

print(classification_report(y_test, y_pred))
```

```python
df = pd.read_csv("/content/IMDB Dataset.csv")
df.head()
```

```python
df['sentiment'].value_counts()
```

```python
df['sentiment'] = df['sentiment'].map({'positive':1,'negative':0})
df.head()
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(df['review'],df['sentiment'], stratify=df['sentiment'])
```

```python
from sklearn.feature_extraction.text import CountVectorizer
v = CountVectorizer()
X_train_count = v.fit_transform(X_train.values)
X_train_count
```

```python
from sklearn.ensemble import RandomForestClassifier

model = RandomForestClassifier(n_estimators=50,criterion='entropy')
model.fit(X_train_count, y_train)
```

```python
X_test_cv = v.transform(X_test)
```

```python
from sklearn.metrics import classification_report

y_pred = model.predict(X_test_cv)

print(classification_report(y_test, y_pred))
```

```python
from sklearn.neighbors import KNeighborsClassifier

model = KNeighborsClassifier(n_neighbors=10,metric='euclidean')

model.fit(X_train_count, y_train)
```

```python
from sklearn.metrics import classification_report

y_pred = model.predict(X_test_cv)

print(classification_report(y_test, y_pred))
```
