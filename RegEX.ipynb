#Meta Characters
Metacharacters are special characters in regular expressions used to define search patterns. \\,[],^,$,.,?,*,+,{},()

```python
#The backslash (\) makes sure that the character is not treated in a special way. If you want to search for the dot(.) in the string then you will find that dot(.) will be treated as a special character as is
#one of the metacharacters (as shown in the above table). So for this case, we will use the backslash(\) just before the dot(.) so that it will lose its specialty.

t = "geeks.forgeeks"
print(re.search(r'.',t))
print(re.search(r'\.',t))
```

```python
# Square Brackets ([]) represent a character class consisting of a set of characters that we wish to match. For example, the character class [abc] will match any single a, b, or c.
#[^0-3] means any character except 0, 1, 2, or 3

string = "The quick brown fox jumps over the lazy dog"
pattern = r"[a-m]"
result = re.findall(pattern, string)

print(result)
```

```python
#Caret (^) symbol matches the beginning of the string i.e. checks whether the string starts with the given character(s) or not.

t = ['The quick brown fox', 'The lazy dog', 'A quick brown fox']
p = r"^The"

for i in t:
  if re.match(p,i):
    print(i,": Match")
  else:
    print(i,": No Match")
```

```python
#{m, n} Braces match any repetitions preceding regex from m to n both inclusive.
#a{2, 4} will be matched for the string aaab, baaaac, gaad, but will not be matched for strings like abc, bc because there is only one a or no a in both the cases.
text = "aaab, baaaac, gaad, abc, bc"
pattern = r"a{2, 4}"

print(re.findall(pattern,text))
```

```python
import re
text = "ababab"
pattern = r"ab"
print(re.findall(pattern, text))
```

```python
text = "aaab, baaaac, gaad, abc, bc"
pattern = r"a{2,4}" ## No space after comma

#{2} means that the preceding character is to be repeated 2 times
#{min,} means the preceding character is matches min or  more  times.
#{min,max} means that the preceding character is repeated at least min & at most max times.

print(re.findall(pattern,text))
```

```python
#Dollar($) symbol matches the end of the string i.e checks whether the string ends with the given character(s) or not.
import re

string = "Hello World!"
pattern = r"World!$"

match = re.search(pattern, string)
if match:
    print("Match found!")
else:
    print("Match not found.")
```

```python
emails =["abc-@mail.com","abc.def@mail.c","abc..def@mail.com","Ahahaaahash","AAHAHoooaa"]

pattern = r"^([a-zA-Z0-9_\-\.]+)@([a-zA-Z0-9_\-\.]+)\.([a-zA-Z]{2,5})$"

for email in emails:
  if re.match(pattern,email):
    print(email," is email address")
  else:
    print(email," not email address")
```
