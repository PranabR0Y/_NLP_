```python
import tensorflow_hub as hub
import tensorflow_text as text
import pandas as pd
from sklearn.utils import resample
import tensorflow as tf
```

```python
df = pd.read_csv("/content/ecommerceDataset.csv")
df.head(6)
```

```python
df['labels'].value_counts(),df.isnull().sum()
```

```python
df['labels'] = df['labels'].map({
    'Household':0,
    'Books':1,
    'Electronics':2,
    'Clothing & Accessories':3

})
df.head()
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(df['Text'],df['labels'], stratify=df['labels'],random_state=42,test_size=0.2)
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.pipeline import Pipeline
from sklearn.neighbors import KNeighborsClassifier

# Fill any remaining NaNs in X_train and X_test
X_train = X_train.fillna('')
X_test = X_test.fillna('')

clf = Pipeline([
    ('vectorizer',TfidfVectorizer()),
    ('KNN',KNeighborsClassifier())
])

clf.fit(X_train,y_train)
```

```python
from sklearn.metrics import classification_report
y_pred = clf.predict(X_test)
print(classification_report(y_test,y_pred))
```

```python
df = pd.read_csv('/content/IMDB Dataset.csv')
df.head()
```

```python
df['sentiment'] = df['sentiment'].map({
    'positive':1,
    'negative':0
})
df.head()
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(df['review'],df['sentiment'], stratify=df['sentiment'],random_state=42,test_size=0.2)
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer

X_train = X_train.fillna('')
X_test = X_test.fillna('')
vectorizer = TfidfVectorizer()
X_train_vectorized = vectorizer.fit_transform(X_train)
X_test_vectorized = vectorizer.transform(X_test)
```

```python
from sklearn.ensemble import RandomForestClassifier
model = RandomForestClassifier(n_estimators=50,criterion='entropy')
model.fit(X_train_vectorized, y_train)
```

```python
from sklearn.metrics import classification_report
y_pred = model.predict(X_test_vectorized)
print(classification_report(y_test,y_pred))
```
